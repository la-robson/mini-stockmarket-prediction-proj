# Predicting daily temperature using LSTM
## Result Visualisation
This contains plots visualising the performance of the LSTM model


## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
print('Libraries imported!')
```

## Import data

```python
loss_filepath = './data/LSTM_loss.csv'
pred_filepath = './data/LSTM_pred.csv'

loss_df = pd.read_csv(loss_filepath)
pred_df = pd.read_csv(pred_filepath)
```

## Plot losses

```python
# Plot the training losses
sns.set_style("ticks")
plt.figure(figsize=(25,5))

plt.plot(loss_df.train_loss)
plt.plot(loss_df.val_loss)

plt.title('Training losses')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

## Plot prediction

```python
# visualise result
sns.set_style("ticks")
plt.figure(figsize=(25,5))
plt.plot(pred_df.temp_val, label = "Actual")
plt.plot(pred_df.temp_pred, label = "Predicted")
plt.show()
```
